```python
# serial date (YYYYMMDD)
year = '2022'
# could use datetime to get between the written and numerical month
month_str, month_int = 'aug', '08'
day = '18'

# path to shape file
mask_path = '/Users/ccroberts/Applications/arcticGIS/sea_ice_masks/site_masks.shp'
site_path = '/Users/ccroberts/Applications/arcticGIS/sea_ice_masks/site_masks.shp'

#########################
date = f'{year}08{day}'
# http url to sea ice data
url = f'https://data.seaice.uni-bremen.de/amsr2/asi_daygrid_swath/n3125/{year}/{month_str}/NorthWestPassage/asi-AMSR2-n3125-{date}-v5.4.tif'
```

```python
def reproject_raster(src, target_crs):
    '''
    Change crs of imported data
    '''
    
    src_crs = src.crs
    src_transform = src.transform
    src_width = src.width
    src_height = src.height

    # Define the target CRS
    target_crs = 'EPSG:3559'

    # Reproject the raster data to the target CRS
    reprojected_data, dst_transform = rs.warp.reproject(
        source=rs.band(src, 1),
        src_transform=src_transform,
        src_crs=src_crs,
        dst_crs=target_crs,
        resampling=rs.enums.Resampling.nearest)
    
    return reprojected_data, dst_transform, target_crs
```

```python
# read sea ice data directly and reproject into EPSG:3559
# transform refers to the affine transformation matrix
# which transforms the pixel indeces to coordinates

# response 200 means success
# response 401 is access denied (e.g., login needed)
response = requests.get(url)
today_index_raw = rs.open(BytesIO(response.content))
today_index_3559, transform_3559, crs_3995 = reproject_raster(today_index_raw, today_index_raw.transform)
```

```python
# import the shape files for the sites
site_gpd = gpd.read_file(site_path)
mask_gpd = gpd.read_file(mask_path)

# Match the projections to the sea ice
mask_gpd = mask_gpd.to_crs('EPSG:3413')

# Grab the important stuff
canarc_mi = site_gpd.iloc[0]
canarc_pi = site_gpd.iloc[1]

mi_mask = mask_gpd.iloc[0].geometry
pi_mask = mask_gpd.iloc[1].geometry
```

```python
masked_mi, transform_mi = mask(today_index_raw, [mi_mask], pad=True, crop=True)
masked_pi, transform_pi = mask(today_index_raw, [pi_mask], pad=True, crop=True)
masked_all, transform_all = mask(today_index_raw, [mi_mask, pi_mask], pad=True, crop=True)
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
show(masked_all, transform=transform_all, cmap='viridis', ax=ax)
```
